Voy a crear una función DEF para automatizar la solicitud del proyecto que requiero

```python
#Necesito Def y retur.
#1.define the paramaters:'Project'
#2.Filter only one project by a variable
def project_chosen (name_project):
    chart_v=alt.Chart(df_complete[df_complete['Project']== name_project]).mark_bar().encode(
        alt.X('Control Period:N'),
        alt.Y ('Weather or number of rainy days:Q'),
        alt.Color('Project')
    )
    return chart_v
project_chosen('A')
project_chosen('B')
# The example has one variable but if i need more.
```

PODEMOS INCLUIR VARIOS PARAMETROS COMO OPACIDAD EN NUESTROS GRÁFICOS.
Los podemos vincular o unir con bind a ciertos métodos interactivos como sliders y demás.

```python
slider= alt.binding_range(max=1, min=0.1,name='Rango de Opacidad',step=0.1)
opacidad= alt.param(value=0.5 ,bind=slider)


chart_v=alt.Chart(df_complete[:42]).mark_bar(opacity=opacidad).encode(
        alt.X('Control Period:N'),
        alt.Y ('Weather or number of rainy days:Q'),
        alt.Color('Project'),
        alt.Column('Project',title='')
         ).add_params(opacidad)
 #
chart_v
```
